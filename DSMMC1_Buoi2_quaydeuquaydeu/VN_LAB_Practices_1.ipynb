-	Nếu 8.0<= điểm < 9.0 - Loại giỏi
-	Nếu 7.0<= điểm < 8.0 - Loại khá

-	Nếu 6.0 <= điểm < 7.0 - Loại trung bình khá
-	Nếu 5.0 <= điểm < 6.0 - Loại trung bình
-	Nếu điểm < 5.0 - Loại kém
Với điểm = ( (điểm toán)*2 + điểm lý + điểm hóa )/4

Viết Chương trình:
a. Nhập vào điểm 3 môn học
b. Tính điểm trung bình
c. Xác định học lực của học sinh dựa trên điểm trung bình

```python
def xac_dinh_so_nguyen_to(n):

    if n < 2:
        print(f"Không có số nguyên tố nhỏ hơn {n}")
    elif n == 2:
        print(f"Số nguyên tố nhỏ hơn {n} là 1")        
    else:

        so_nguyen_to = [1]
        for number in range(2,n):
            so_du = 1
            for thua_so in range(2,number):
                # print(f'1. thua_so {thua_so}, so_du {so_du}')
                so_du = so_du * (number % thua_so)
            if so_du !=0:
                so_nguyen_to.append(number)
        print(f'Các số nguyên tố nhỏ hơn {n} là {so_nguyen_to}')
xac_dinh_so_nguyen_to(30)
```

Viết chương trình liệt kê n số nguyên tố đầu tiên.


```python
def liet_ke_SNT(m):
    list_snt = [1]
    so_dem = 1
    snt = 1

    while so_dem < m:
        so_du = 1
        snt += 1
        for number in range(2,snt):
            so_du = so_du * (snt % number)
        
        if so_du !=0:
            so_dem += 1
            list_snt.append(snt)
    print(f'5. {m} số nguyên tố đầu tiên là {list_snt}')
liet_ke_SNT(15)
```

Question 11:
Viết chương trình thực hiện nhập một xâu ký tự
a) Tìm từ dài nhất trong xâu đó.
b) Từ đó xuất hiện ở vị trí nào?
(Chú ý. nếu có nhiều từ có độ dài giống nhau thì chọn từ đầu tiên tìm thấy).


```python
def tim_chu_dai(chuoi):
    std = chuoi.strip().split(' ')
    n = 0
    vi_tri = 0
    chu_dai_nhat = ""
    for text in std:
        if n < len(text):
            n = len(text)
            chu_dai_nhat = text
            vi_tri = chuoi.find(chu_dai_nhat)

    return print(f'Chữ dài nhất là {chu_dai_nhat}, có {n} ký tự, đứng ở vị trí {vi_tri}')

chuoi = 'nguyen Trong Phuc that deptraiqua troi '
tim_chu_dai(chuoi)
```
